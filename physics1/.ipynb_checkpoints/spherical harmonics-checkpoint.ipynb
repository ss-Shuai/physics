不同的书上有不同的约定，咱还是用 Condon-Shortley 相位约定，m≥0时，定义如下：

$Y_{n}^{m}(\theta,\phi)\equiv\epsilon\sqrt{\frac{2n-1}{4\pi}\frac{(n-|m|)!}{(n+|m|)!}} \quad  $
$P_{h}^{|m|}(\cos\theta)e^{i m\phi},\quad \epsilon=\left\{\begin{array}{l l}{{(-1)^{m},}}&{{~~~m\geq0,}}\\ {{1,}}&{{~~~e l s e}}\end{array}\right.$

其中，$(-1)^m$即Condon-Shortley相因子，大概是为了方便做角动量代数的，连带勒让德函数定义为（$Rodriguez$公式）：

$P_n^{m}(x)\equiv\frac{1}{2^{n}n!}(1-x^{2})^{m/2}\frac{d^{ n+m}}{d x^{n+m}}\quad\!(x^{2}-1)^{n}.$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
import mpl_toolkits.mplot3d.axes3d as axes3d
```

## scipy.special.sph_harm(m, n, theta, phi) = <ufunc 'sph_harm'>¶

$Y_{n}^{m}(\theta,\phi)=\sqrt{\frac{2n+1}{4\pi}\frac{(n-m)!}{(n+m)!}} \quad e^{i m\theta}  \qquad P_{n}^{m}(\cos(\phi))$

### Parameters:	

#### m : array_like

Order of the harmonic (int); must have |m| <= n.

#### n : array_like

Degree of the harmonic (int); must have n >= 0. This is often denoted by l (lower case L) in descriptions of spherical harmonics.

#### theta : array_like

Azimuthal (longitudinal) coordinate; must be in [0, 2*pi].

#### phi : array_like

Polar (colatitudinal) coordinate; must be in [0, pi].

## scipy.special.lpmv(m, v, x) = <ufunc 'lpmv'>

$P_{v}^{m}=(-1)^{m}(1-x^{2})^{m/2}\frac{d^{\,m\,}}{d x^{\,m\,}}P_{v}(x)$

$P_{v}=\sum_{k=0}^{\infty}{\frac{(-v)_{k\,}(v+1)_{k\,}}{(k!)^{2}}}\, \left({\frac{1-x}{2}}\right)^{k}$

```python
theta, phi = np.linspace(0, np.pi, 100), np.linspace(0, 2*np.pi, 100)
THETA, PHI = np.meshgrid(theta, phi)
#help(special.sph_harm)
R = (special.sph_harm(2,3,PHI,THETA).real)**2
#m=2, l=3
X = R * np.sin(THETA) * np.cos(PHI)
Y = R * np.sin(THETA) * np.sin(PHI)
Z = R * np.cos(THETA)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')
plot = ax.plot_surface(
    X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('jet'),
    linewidth=0, antialiased=False, alpha=0.5)

# below are codes copied from stackoverflow, to make the scaling correct
max_range = np.array([X.max()-X.min(), Y.max()-Y.min(), Z.max()-Z.min()]).max() / 2.0
mid_x = (X.max()+X.min()) * 0.5
mid_y = (Y.max()+Y.min()) * 0.5
mid_z = (Z.max()+Z.min()) * 0.5
ax.set_xlim(mid_x - max_range, mid_x + max_range)
ax.set_ylim(mid_y - max_range, mid_y + max_range)
ax.set_zlim(mid_z - max_range, mid_z + max_range)

#ax.view_init(elev=30,azim=0) #调节视角，elev指向上(z方向)旋转的角度，azim指xy平面内旋转的角度

plt.show()
```
